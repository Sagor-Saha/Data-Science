```python
# Loading the packages 
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

# Basic Barplot

```python
# http://python-graph-gallery.com/barplot/
# https://matplotlib.org/tutorials/introductory/pyplot.html 
# Make a fake dataset:
heights = [3,12,4,5,6,7,8]
bars = ('A','B','C','D','E','F','G')
#Create bars
plt.bar(bars, heights)

#Create names on the x-axis
plt.xticks(bars)

#Show graphic
plt.show()
```

# Horizontal Barplot

```python
# Create horizontal bars
plt.bar(bars, heights)
 
# Create names on the y-axis
plt.yticks(bars)
 
# Show graphic
plt.show()
```

# Adjusting the colors

```python
# RGB is a way of making colors.
# color =  red, green and blue + the transparency
# Put transparency = 0 
# Put transparency = 1
plt.bar(bars, heights, color=(0.2, 0.4, 0.6))
plt.xticks(bars)
plt.show()
# Different color for each bar 
# Provide a list of color names to the color argument
plt.bar(bars, heights, color=['black', 'red', 'green', 'blue', 'cyan'])
plt.xticks(bars)
plt.show()
```

```python
# Control color of the border 
# The edgecolor argument allows to color the borders of barplots.
plt.bar(bars, heights, color=(0.1, 0.1, 0.1, 0.1),  edgecolor='red')
plt.xticks(bars)
plt.show()
```

```python
# Create bars and choose color
plt.bar(bars, heights, color = (0.5,0.1,0.5,0.6))
 
# Add title and axis names
plt.title('$Height_{bars}$', fontsize=14)
plt.xlabel('Name', fontsize=14)
plt.ylabel('Height', fontsize=14)
 
# Limits for the Y axis
plt.ylim(0,60)
 
# Create names
plt.xticks(bars)
 
# Show graphic
plt.show()
```

# Basic Line chart

```python
# Create a dataframe 
df = pd.DataFrame({'year': [2015, 2016, 2017, 2018, 2019], 'T20': [24, 36, 12, 18, 25], 'ODI': [89, 24, 37, 63, 98]})

# Draw the plot 
plt.plot( 'year', 'T20', data = df)
plt.show()
```

```python
# Draw the plot 
# https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.legend.html 
plt.plot( 'year', 'T20', data = df, linewidth = 5)
plt.plot('year', 'ODI', data = df, c='red', linewidth = 5)
plt.xticks(np.arange(min(df.year), max(df.year)+1, 1), fontsize = 15)
plt.yticks(np.arange(0,125,25), fontsize = 15)
plt.title('Points scored by India', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Points', fontsize = 15)
plt.legend(loc='upper center')
plt.grid(b=True, alpha=0.9) #which argument to choose the axis 
plt.show()
```

# Violin plot

```python
# https://python-graph-gallery.com/violin-plot/
# Violinplots allow to visualize the distribution of a numeric variable for one or several groups. 
# It is really close from a boxplot, but allows a deeper understanding of the density. 
# Violins are particularly adapted when the amount of data is huge and showing individual observations gets impossible. 
# Violinplots probably deserve more attention compared to boxplot that can sometimes hide features of the data.

# If you have only one numerical variable, you probably better have to make an histogram or a density plot. 
# But you can still use the violinplot function to describe the distribution of this variable, as follow:
# from pylab import *
df = sns.load_dataset('iris')
 
# plot
sns.violinplot(x=df["species"], y=df["sepal_length"])
```

```python
sns.boxplot(x="species", y="sepal_length", data=df)
```

# Using the correlation matrix, peform the following:
Find the correlation matrix for the movies.
Draw a heatmap with the correlation matrix and display the values of correlation coefficients for each pair of features.
Create a mask to remove the duplicate values from the heatmap.

```python
corr_values = movies.corr()
plt.figure(figsize = (15,10))
sns.heatmap(corr_values, vmin = -1, cmap="coolwarm", annot=True)
plt.show()
```

```python
# Creating a mask for removing duplicates 
mask = np.zeros_like(movies.corr(), dtype=np.bool) # create matrix of 0's of the given matrix site
mask[np.triu_indices_from(mask)] = True # changing the lower triangle matrix element to True (1)
plt.figure(figsize = (15,10))
sns.heatmap(corr_values, vmin = -1, cmap="coolwarm", annot=True, mask=mask)
plt.show()
```
